```python
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup as bs
```

```python
price_list = []
url_list = []
products_df = pd.read_csv('product-features.csv')
link_list = list(products_df['link'])

products_df = pd.read_csv('product-features.csv')
products_df
```

```python
headers = {'User-agent': 'Mozilla/5.0'}

for url in  link_list:

    page = requests.get(url, headers=headers)
    bs_page = bs(page.content, 'lxml')

    try: 
        table_values = bs_page.find('b', class_="css-0")
        print(table_values)
        price = table_values.contents
        price_list.append(price)
        url_list.append(url)
        
    except:
        print(url)

    time.sleep(1800)
```

```python
dict = {"price" : price_list, "url":url_list }
price_df = pd.DataFrame(dict)
price_df.to_csv('price.csv')
```
